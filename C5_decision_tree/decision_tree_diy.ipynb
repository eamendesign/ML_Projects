### Purpose of this function

This function is used to calculate:

* How "chaotic" the labels in a dataset are.
The formula is:
$$H(S) = -\sum p_i \log_2 p_i$$
If a set contains:
* Only cats → Entropy = 0
* Half cats and half dogs → Maximum entropy
</div>

```python
def entropy(data):
    """
    计算数据集标签的熵
    H(S) = - sum(p_i * log2(p_i))
    """
    labels = []
    for i in range(len(data)):
        row = data[i]
        label = row["label"]
        # print(f"label: {label}")
        labels.append(label)

    count = Counter(labels) #  Counter({'cat': 3, 'dog': 2})
    # print(count)
    total = len(labels)

    ent = 0.0
    for c in count.values():
        
        p = c / total
        # print(f"c = {c}, total = {total}, p = {p}, p * log2(p) = {p * math.log2(p)}")
        ent -= p * math.log2(p)

    return ent
```

```python
example_all_cat = [
    {"label": "cat"},
    {"label": "cat"},
    {"label": "cat"}
]

print("全是猫时的熵 H(S) = - sum(p * log2(p)) :", entropy(example_all_cat))
```

```python
def split_dataset(data, feature):
    """
    按某个特征切分数据
    返回一个字典:
    {
        特征值1: [样本1, 样本2, ...],
        特征值2: [样本3, 样本4, ...]
    }
    """
    groups = {} # 新建 group 字典

    for row in data:

        key = row[feature]  # feature: sound, key: meow, bark
        # print(f"feature: {feature} | key: {key}")

        if key not in groups:   # 检查 meow 或者 bark（是否存在字典）
            groups[key] = []    #  没有？创建它！

        groups[key].append(row) # 把row放进对应key的字典里

    return groups
```

### Information Gain Formula

$$IG(S, A) = H(S) - \sum \frac{|S_v|}{|S|} H(S_v)$$

**Meaning:**

* Before splitting, chaotic the data is $ H(S)$.
* After splitting, the average remaining chaos is $\sum \frac{|S_v|}{|S|} H(S_v)$.
* The reduction is the Information Gain.

</div>

```python
def information_gain(data, feature):
    """
    计算某个特征的信息增益
    IG(S, A) = H(S) - sum( |Sv|/|S| * H(Sv) )
    """
    # 1. 分裂前的熵
    before_split_entropy = entropy(data)

    # 2. 按特征切分
    groups = split_dataset(data, feature)
    total_samples = len(data)

    # 3. 分裂后的加权平均熵
    after_split_entropy = 0.0
    for group_name, group_data in groups.items():
        group_weight = len(group_data) / total_samples
        group_entropy = entropy(group_data)
        after_split_entropy += group_weight * group_entropy

    # 4. 信息增益 = 分裂前熵 - 分裂后熵
    info_gain = before_split_entropy - after_split_entropy

    return info_gain
```

```python
def information_gain_verbose(data, feature):
    """
    带详细打印的信息增益计算
    """
    before_split_entropy = entropy(data)

    total_samples = len(data)

    print(f"分裂前熵 H(S) = - sum(p * log2(p)) : {before_split_entropy:.4f}")
    
    print(f"\n========== 特征: {feature} 分裂后结果 ==========")
    
    groups = split_dataset(data, feature)
    
    for k, v in groups.items():
        print(f"{k}: {len(v)} 个样本")
        # for item in v:
        #     print(item)

    after_split_entropy = 0.0
    for group_name, group_data in groups.items():

        print(f" ==== 特征值 = {group_name} =====")
        group_entropy = entropy(group_data)
        print(f"  子集熵 = {group_entropy:.4f}")
        
        print(f"  样本数 = {len(group_data)}")
        print(f"  总样本数 = {total_samples}")
        group_weight = len(group_data) / total_samples
        print(f"  权重 = 样本数/总样本数  = {group_weight:.4f}")
        
        after_each_split_entropy = group_weight * group_entropy
        print(f" 特征值{group_name} 的加权熵 = 权重 * 子集熵 = {after_each_split_entropy:.4f}")
        after_split_entropy += group_weight * group_entropy
        

    print(f"*******分裂后 sum( |Sv|/|S| * H(Sv) ) 总加权熵: {after_split_entropy:.4f}")

    info_gain = before_split_entropy - after_split_entropy
    print(f"*******信息增益 IG(S, A) = preH(S) - posH(S): {info_gain:.4f}")

    return info_gain
```

```python
def best_feature(data, feature_list):
    """
    在所有候选特征中，找到信息增益最大的那个
    """
    best_f = None
    best_ig = -1

    for f in feature_list:
        ig = information_gain(data, f)
        if ig > best_ig:
            best_ig = ig
            best_f = f

    return best_f, best_ig
```

```python
best_f, best_ig = best_feature(dataset, features)
print("最优特征:", best_f)
print("对应的信息增益:", best_ig)
```

```python
def all_same_label(data):
    """
    判断数据集里的标签是否全都相同
    """
    # 1. 取出第一个样本的标签
    if len(data) == 0:  # 空数据集
        return True
    
    first_label = data[0]["label"]
    
    # 2. 遍历所有样本，检查是否都和第一个相同
    for row in data:
        if row["label"] != first_label:
            return False  # 发现不同的标签，返回False
    
    return True  # 全都相同
```

```python
def majority_label(data):
    """
    返回数据集中出现次数最多的标签
    """
    labels = [row["label"] for row in data]
    return Counter(labels).most_common(1)[0][0]
```

```python
example3 = [
    {"label": "cat"},
    {"label": "cat"},
    {"label": "dog"}
]

print("多数类是:", majority_label(example3))
```

```python
def build_tree(data, feature_list):
    """
    递归构建 ID3 风格的决策树
    """
    # 情况1：当前节点已经纯了
    if all_same_label(data):
        return {"label": data[0]["label"]}

    # 情况2：没有可用特征了
    if len(feature_list) == 0:
        return {"label": majority_label(data)}

    # 情况3：选择最优特征
    best_f, best_ig = best_feature(data, feature_list)

    tree = {
        "feature": best_f,
        "info_gain": best_ig,
        "children": {},
        "majority_label": majority_label(data)
    }

    groups = split_dataset(data, best_f)
    
    remaining_features = []  # 创建一个空列表，用来存放剩余的特征

    for feature in feature_list:  # 遍历所有特征
        if feature != best_f:      # 如果这个特征不是刚刚用掉的最佳特征
            remaining_features.append(feature)  # 就把它加入剩余特征列表

    print(f"已使用特征: {best_f}")
    print(f"剩余可用特征: {remaining_features}")

    # 遍历每一个特征值对应的子数据集
    for feature_value in groups:  # feature_value 是特征值，如 "红色"、"黄色"等
        subset = groups[feature_value]  # 获取这个特征值对应的子数据集
        
        print(f"\n 正在处理 {best_f} 特征值 '{feature_value}'")
        print(f"  该子集有 {len(subset)} 个样本")
        
        # 递归调用：用子数据集和剩余特征继续构建树
        child_tree = build_tree(subset, remaining_features)
        
        # 把构建好的子树挂在当前节点的 children 下
        tree["children"][feature_value] = child_tree
        
        print(f"  ✓ 子树构建完成，已挂载到 tree['children']['{feature_value}']")

    return tree
```

```python
tree = build_tree(dataset, features)
tree
```

```python
def print_tree(tree, indent=""):
    """
    把树结构漂亮地打印出来
    """
    if "label" in tree:
        print(indent + "-> Predict:", tree["label"])
        return

    print(indent + f"[Feature: {tree['feature']}, Info Gain: {tree['info_gain']:.4f}]")
    for value, subtree in tree["children"].items():
        print(indent + f"  If {tree['feature']} == {value}:")
        print_tree(subtree, indent + "    ")
```

```python
def predict(tree, sample):
    """
    用训练好的树预测一个新样本
    如果遇到训练中没见过的特征值，就返回当前节点多数类
    """
    if "label" in tree:
        return tree["label"]

    feature = tree["feature"]
    value = sample.get(feature)

    if value in tree["children"]:
        return predict(tree["children"][value], sample)
    else:
        return tree["majority_label"]
```
